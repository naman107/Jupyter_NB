### Histograms

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon = pd.read_csv('C:/Users/naman.agarwal/Desktop/pokemon.csv')
```

```python
pokemon.head(3)
```

```python
plt.figure(figsize=(15,8))
plt.hist(pokemon['Attack'],bins=100)
plt.show()
```

```python
plt.figure(figsize=(15,8))
bins = [0,1,50,100,200,500]
plt.hist(pokemon['Attack'],bins=bins)
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.hist(pokemon['Attack'],bins=100,density=True,orientation='horizontal')
plt.show()
```

```python
plt.figure(figsize=(15,8))
color = ['red']
plt.hist(pokemon['Attack'],bins=100,density=True,log=True,color=color)
plt.show()
```
