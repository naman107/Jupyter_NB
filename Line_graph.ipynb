```python
plt.figure(figsize=(12,5))
plt.plot(years,y1,label='Canada',marker='o',markerfacecolor='b',markersize=8,linestyle='--')
plt.plot(years,y2,label='Australia',marker='o',markerfacecolor='orange',markersize=8)
plt.title('Gas Prices',fontdict={'fontweight':'bold','fontsize':15})
plt.xticks(years)
plt.xlabel('Years',fontdict={'fontweight':'bold','fontsize':13})
plt.ylabel('USD',fontdict={'fontweight':'bold','fontsize':13})
plt.legend()
plt.show()
```

```python
countries = ['Australia','Canada','France','Germany','Italy','Japan','Mexico','South Korea','UK','USA']
gas.fillna(0)
```

```python
plt.figure(figsize=(14,8))
for x in countries:
    plt.plot(years,gas[x],label=x,marker='o',markersize=7,linewidth=2)
plt.title('Gas Prices',fontdict={'fontweight':'bold','fontsize':15})
plt.xticks(years)
plt.xlabel('Years',fontdict={'fontweight':'bold','fontsize':13})
plt.ylabel('USD',fontdict={'fontweight':'bold','fontsize':13})
plt.rc('legend',fontsize=15)
plt.legend(bbox_to_anchor=(1, 1.05))
plt.show()
```
